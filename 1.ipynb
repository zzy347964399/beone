```python
import pandas as pd
DATA_DIR = "/Users/pickle/.cache/kagglehub/datasets/asjad99/mimiciii/versions/1/mimic-iii-clinical-database-demo-1.4"

# 图 3：看 ICU LOS 和 first_careunit 的真实数据
icu = pd.read_csv(f"{DATA_DIR}/ICUSTAYS.csv", usecols=["first_careunit","los"])
print(icu[icu["los"]>0].groupby("first_careunit")["los"].describe())

# 图 4：看实际死亡率数值
adm = pd.read_csv(f"{DATA_DIR}/ADMISSIONS.csv", usecols=["admission_type","hospital_expire_flag"])
adm["hospital_expire_flag"] = adm["hospital_expire_flag"].fillna(0).astype(int)
g = adm.groupby("admission_type")["hospital_expire_flag"].agg(total="count", dead="sum")
g["rate"] = g["dead"]/g["total"]*100
print(g)
```
